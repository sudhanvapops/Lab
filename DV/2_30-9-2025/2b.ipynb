### 2b Develop a python program to convert binary to decimal, octal to hexadecimal using functions.


```python
def bin2Dec(val):
    return int(val,2)

def oct2hex(val):
    bin = int(val,8)
    hexDec = hex(bin)
    return hexDec

try:
    num1 = input("Enter a Binary Number: ")
    print(f"Binary {num1} → Decimal {bin2Dec(num1)}")

    num2 = input("Enter a Octal Number: ")
    print(f"Octal {num2} → HexaDecimal {oct2hex(num2)}")
except ValueError:
    print(f"Invalid Input Type")
```
