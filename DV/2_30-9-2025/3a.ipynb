### 3a. Write a Python program that accepts a sentence and find the number of words, digits, uppercase letters and lowercase letters

```python
import string
```

```python
digitCount = upperCaseCount = lowerCaseCount = 0
```

```python
sentence = input("Enter a Sentence: ")
wordList = sentence.strip().split()
```

```python
print(f"This Sentence has {len(wordList)} words")
```

```python
for ch in sentence:
    if ch in string.digits:
        digitCount += 1
    elif ch in string.ascii_uppercase:
        upperCaseCount += 1
    elif ch in string.ascii_lowercase:
        lowerCaseCount += 1
```

```python
print(f"This sentence has\n{digitCount} digits\n{upperCaseCount} UpperCase letters\n{lowerCaseCount} lowercase letters")
```
