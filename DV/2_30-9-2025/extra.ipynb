### Test For Prime Number

```python
import math
```

```python
def isPrime(val):

    if val <= 1:
        return False

    if val == 2:
        return True
    
    if val % 2 == 0:
        return False

    for i in range(3,int(math.sqrt(val))+1,2):
    # for i in range(3,val,2):
        if val % i == 0:
            return False
        
    return True
```

```python
print(f"{7} is prime: {isPrime(7)}")
print(f"{11} is prime: {isPrime(11)}")
print(f"{21} is prime: {isPrime(21)}")
print(f"{49} is prime: {isPrime(49)}")
print(f"{-1} is prime: {isPrime(-1)}")
```

```python
a = 56
b = 112

gcd = math.gcd(a, b)
lcm = math.lcm(a, b)  

print(f"GCD of {a} and {b} is {gcd}")
print(f"LCM of {a} and {b} is {lcm}")

ngcd,nlcm = 0,0
for i in range(1,min(a,b)+1):
    if (a % i == 0 and b % i == 0):
        ngcd = i

nlcm = (a * b) // ngcd
print(f"GCD of {a} and {b} is {ngcd}")
print(f"GCD of {a} and {b} is {nlcm}")
```
