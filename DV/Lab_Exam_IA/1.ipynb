```python
marks = []
for i in range(3):
    m = int(input(f"Input Marks: {i+1}"))
    marks.append(m)

avgMarks = sum(marks) / 3
bestOfTwo = sorted(marks,reverse=True)[:2]

print(f"Avg Marks: {avgMarks}")
print(f"Best of Marks: {bestOfTwo}")
```

```python
pal = input("Enter a number: ")
if (pal == pal[::-1]):
    print(f"{pal} Its a Palindrone")
else:
    print(f"{pal} Its not a palindrome")

from collections import Counter

c = Counter(pal)
for k,i in c.items():
    print(f"{k} Occurs {i} times")
```

```python
def F(n):
    if n == 0: return 0
    if n == 1: return 1

    return F(n-1) + F(n - 2)


try:
    n = int(input("Enter n: "))
    if (n>0):
        print(F(10))
    else:
        print("N should be greate than 0")
except ValueError:
    print("Enter a Number")
```

```python
def bin2Dec(inp):
    return int(inp,2)

def oct2hex(octVal):
    dec = int(octVal,8)
    return hex(dec)

try:
    binNum = input("Enter a Binary Number: ")
    print(bin2Dec(binNum))
    ocyNum = input("Enter a Octal Number: ")
    print(oct2hex(ocyNum))
except ValueError:
    print("Invalid Input Type")
```

```python
import string

upperCase = lowercase = digits = 0

inpString = input("Enter a String: ").strip()
wordlist = inpString.split()


print(f"Words: {len(wordlist)}")

for ch in inpString:
    if ch in string.digits:
        digits+=1
    elif ch in string.ascii_uppercase:
        upperCase += 1
    elif ch in string.ascii_lowercase:
        lowercase += 1

print(f"This sentence has\n{digits} digits\n{upperCase} UpperCase letters\n{lowercase} lowercase letters")
```
