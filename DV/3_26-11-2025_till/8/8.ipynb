```python
# Depricated
# p2.square(x, 3*y, legend_label="3*sin(x)", fill_color=None, line_color="green")

# Circle also depricated

import numpy as np
from bokeh.io import output_notebook
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show

output_notebook()

x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

TOOLS = "pan, wheel_zoom, box_zoom, reset, save, box_select"

p1 = figure(title="Example 1", tools=TOOLS)

# Use scatter instead of circle, with marker="circle"
p1.scatter(x, y, marker="circle", size=6, legend_label="sin(x)")
p1.scatter(x, 2*y, marker="circle", size=6, legend_label="2*sin(x)", color="orange")
p1.scatter(x, 3*y, marker="circle", size=6, legend_label="3*sin(x)", color="green")
p1.legend.title = 'Markers'

p2 = figure(title="Example 2", tools=TOOLS)

# scatter with circle marker
p2.scatter(x, y, size=6, marker="circle", legend_label="sin(x)")
p2.line(x, y, legend_label="sin(x)")

p2.line(
    x, 2*y,
    legend_label="2*sin(x)",
    line_dash=(4, 4),
    line_color="orange",
    line_width=2
)

# scatter with square marker
p2.scatter(
    x, 3*y,
    marker="square",
    size=8,
    legend_label="3*sin(x)",
    fill_color=None,
    line_color="green",
)

p2.line(x, 3*y, legend_label="3*sin(x)", line_color="green")
p2.legend.title = 'Lines'

show(gridplot([p1, p2], ncols=2, width=400, height=400))

```
