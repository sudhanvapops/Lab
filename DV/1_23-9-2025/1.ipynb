### 1A Program to find best out of two marks Average out of three test from users

```python
marks = []

for i in range(3):
    mark = input(f"Type Marks of subject: {i+1}")
    marks.append(int(mark))

bestOfTwo = sorted(marks,reverse=True)[:2]
averageMarks = sum(bestOfTwo) / 2

print(f"The average of two marks: {averageMarks}")
```

### 1B Develop a Python program to check whether a given number is palindrome or not 
### and also count the number of occurrences of each digit in the input number. 

```python
num = input("Enter a Number: ")
if (num == num[::-1]):
    print(f"Number: {num} is a Palindrome")
else: 
    print(f"Number {num} is not a palindrome")


occur = {}
for i in num:
    occur[i] = occur.get(i,0) + 1

for key in occur:
    print(f"{key} appeared {occur[key]} times")
```

### 1B Another Way in Lab Manual

```python
from collections import Counter
counter = Counter(num)
for key in counter:
    print(f"{key} appeared {counter[key]} times")
```
