### 2A Defined as a function F as Fn = Fn-1 + Fn-2. 
### Write a Python program which accepts a value for N (where N >0) as input and pass this value to the function.
### Display suitable error message if the condition for input value is not followed.

```python
def F(n):

    if(n == 0): return 0
    if(n==1): return 1

    return F(n-1) + F(n-2)

# Cant do without else it wont take exit(1)
try:
    n = int(input("Enter N: "))
    if (n<0):
        print(f"Input {n} should be greater than 0")
    else:
        print(f"F({n}) = {F(n)}")
except ValueError:
    print("Try with numeric value")
```
